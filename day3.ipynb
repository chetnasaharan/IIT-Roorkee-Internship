```python
def fibonacci(n):
    if n <= 1:   
        return n                                             # first two terms of the series are 0 and 1
    else:
        return fibonacci(n - 1) + fibonacci(n - 2)           # calcluate nth number using (n-1)th and (n-2)th number
a = int(input("Enter number of terms: ")) 
print(a, "terms of the Fibonacci series:")                      
series = [fibonacci(i) for i in range(a)]
print(series)
```

```python
def quiz():
    questions = {1,2,3,4,5}
    score = 0
    for question in questions: 
        if question == 1:
            print("What is the capital of France?")
            print("a) London")
            print("b) Paris")
            print("c) Rome")
            print('d) Berlin')
            answer = input("Enter your answer: ")
            print(answer)
            if answer.lower() == "b":
                print("Correct!")
                score+=2
            else:
                print("Incorrect. The correct answer is b) Paris.")
        elif question == 2:
            print("What is the largest planet in our solar system?")
            print("a) Earth")
            print("b) Mars")
            print("c) Jupiter")
            print('d) Saturn')
            answer = input("Enter your answer: ")
            print(answer)
            if answer.lower() == "c":
                print("Correct!")
                score+=2
            else:
                print("Incorrect. The correct answer is c) Jupiter.")
        elif question == 3:
            print("What is the chemical symbol for water?")
            print("a) H2O")
            print("b) CO2")
            print("c) O2")
            print('d) NaCl')
            answer = input("Enter your answer: ")
            print(answer)
            if answer.lower() == "a":
                print("Correct!")
                score+=2
            else:
                print("Incorrect. The correct answer is a) H2O.")
        elif question == 4:
            print("What is the capital of Japan?")
            print("a) Seoul")
            print("b) Beijing")
            print("c) Tokyo")
            print('d) Kyoto')
            answer = input("Enter your answer: ")
            print(answer)
            if answer.lower() == "c":
                print("Correct!")
                score+=2
            else:
                print("Incorrect. The correct answer is c) Tokyo.")
        elif question == 5:
            print("What is the largest mammal?")
            print("a) Elephant")
            print("b) Blue Whale")
            print("c) Giraffe")
            print('d) Hippopotamus')
            answer = input("Enter your answer: ")
            print(answer)
            if answer.lower() == "b":
                print("Correct!")
                score+=2
            else:
                print("Incorrect. The correct answer is b) Blue Whale.")    
    print(f"Your final score is: {score}/10")
```

```python
quiz()
```

```python
import string
import random 
def passwordgen(length):
    characters = string.ascii_letters + string.digits + string.punctuation
    password = ''.join(random.choice(characters) for i in range(length))
    return password
```

```python
print(passwordgen(15))
```

```python
import string
import random
def generate_pass(length):
 return ''.join(random.choice(string.printable) for i in range(length) if random.choice(string.printable) not in string.whitespace)    # to avoid whitespace characters in the password
l = int(input("Enter password length: "))
print("Generated password:", generate_pass(l))
  
```

```python
def fact_it(n):                    #iterative 
    result = 1
    for i in range (1,n+1):
        result *= i 
    return result
def fact_rec(n):               #recursive
    if n==0 or n==1:
        return 1
    else:
        return n*fact_rec(n-1)
```

```python
print(fact_it(5))
print(fact_rec(5))
```
