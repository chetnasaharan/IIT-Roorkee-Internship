1) First make a file and then read a text file and count lines, words, characters 

```python
# Create and write to a text file
with open("sample.txt", "w", encoding="utf-8") as file:
    file.write("MY NAME IS CHETNA\n")
    file.write("MYSELF CHETNA.\n")
    file.write("CHETNA THIS SIDE.")

# Read the file
with open("sample.txt", "r", encoding="utf-8") as file:
    content = file.read()

# Count lines, words, and characters
lines = len(content.splitlines())
words = len(content.split())
characters = len(content)

# Display results
print("File Content:")
print(content)

print("\nStatistics:")
print("Lines:", lines)
print("Words:", words)
print("Characters:", characters)
```

2) Write student data to a CSV file and read it back.

4) Write a safe division function that handles ZeroDivisionError.

```python
def safe_division(a, b):
    try:
        print("Result =", a / b)
    except ZeroDivisionError:
        print("Cannot divide by zero!")

safe_division(20, 4)
safe_division(20, 0)
```

```python
# Create a log file
with open("log.txt", "w") as file:
    file.write("ERROR: File not found\n")
    file.write("WARNING: Low memory\n")
    file.write("ERROR: File not found\n")
    file.write("ERROR: Database connection failed\n")
    file.write("WARNING: Low memory\n")
    file.write("INFO: System started\n")

print("Log file created successfully!")
error_count = {}

with open("log.txt", "r") as file:
    for line in file:
        if line.startswith("ERROR"):
            error_type = line.strip()

            if error_type in error_count:
                error_count[error_type] += 1
            else:
                error_count[error_type] = 1

print("\nError Summary:")
for error, count in error_count.items():
    print(f"{error} -> {count}")
```

```python
import csv

# Create sample marks file
with open("marks.csv", "w", newline="") as file:
    writer = csv.writer(file)

    writer.writerow(["Name", "Math", "Science", "English"])
    writer.writerow(["Aman", 80, 75, 90])
    writer.writerow(["Riya", 60, 70, 65])
    writer.writerow(["John", 85, 88, 92])

print("marks.csv created successfully!")
import csv

def get_grade(avg):
    if avg >= 90:
        return "A+"
    elif avg >= 75:
        return "A"
    elif avg >= 60:
        return "B"
    else:
        return "C"

# Read input file and write output file
with open("marks.csv", "r") as infile, open("result.csv", "w", newline="") as outfile:
    reader = csv.reader(infile)
    writer = csv.writer(outfile)

    header = next(reader)  # skip header
    writer.writerow(header + ["Average", "Grade"])

    for row in reader:
        name = row[0]
        marks = list(map(int, row[1:]))

        avg = sum(marks) / len(marks)
        grade = get_grade(avg)

        writer.writerow(row + [round(avg, 2), grade])

print("result.csv created successfully!")
```

```python
# Create a sample file
with open("data.txt", "w") as file:
    file.write("Python is easy to learn\n")
    file.write("Java is also popular\n")
    file.write("Python is used in AI\n")
    file.write("C++ is powerful\n")

print("File created successfully!")
keyword = input("Enter keyword to search: ")

print("\nMatching lines:\n")

with open("data.txt", "r") as file:
    found = False

    for line in file:
        if keyword.lower() in line.lower():
            print(line.strip())
            found = True

    if not found:
        print("No matching lines found.")
```
