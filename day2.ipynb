```python
import random
Num = random.randrange(1,101)
print ('Guess a number between 1 and 100 ')
guess = int(input('enter your guess: '))
if guess < Num :
    print ('number is higher, try again.')
elif guess > Num :
    print('number is lower, try again.')
else :
    print(f'your guess {guess} is correct')
```

```python
n = int(input('enter the number of rows: '))
for i in range(0,n):
    for j in range(i,n):                       # first empty decreasing triangle
        print(" ",end=" ")
    for j in range (0,i):                      # second triangle , filled and increasing 
        print("*",end=" ")
    for j in range (i+1):                      # third triangle , filled and increasing
        print("*",end=" ")
    print()
```
